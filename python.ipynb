```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
x = np.arange(10)
y = x**2
```

```python
plt.plot(x, y)
plt.title("y=x^2");
```

```python
plt.plot(x, y)
plt.grid(True)
plt.title("y=x^2");
```

```python
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("y=x^2");
```

```python
plt.plot(x, y, "-s")
plt.grid(True)
plt.title("y=x^2");
```

```python
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("y=x^2");
plt.xlabel("X point")
plt.ylabel("Y point")
```

```python
plt.plot(x, y, "-o", lw=3)
plt.grid(True)
plt.title("y=x^2");
plt.xlabel("X point")
plt.ylabel("Y point")
```

```python
plt.plot(x, y, "-o", lw=3)
plt.grid(True)
plt.title("y=x^2");
plt.xlabel("X point")
plt.ylabel("Y point")
plt.xscale("log")
```

```python
img = Image.open("macaque.jpg")
img = np.asarray(img)
plt.imshow(img, extent=[0, 8, 0, 80], aspect="auto")
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("y=x^2")
plt.gca().set(xlim=[-0.5, 9.5], ylim=[-5, 85]);
```
