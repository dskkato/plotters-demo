```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

## プロットする二次関数データ

```python
x = np.arange(10)
y = x**2
```

### 基本的なプロット

```python
plt.plot(x, y)             # curve 1
plt.title("y=x^2");
```

### グリッド

```python
plt.plot(x, y)
plt.grid(True)
plt.title("y=x^2");
```

### 丸いマーカー

```python
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("y=x^2");
```

### 四角のマーカー

```python
plt.plot(x, y, "-s")
plt.grid(True)
plt.title("y=x^2");
```

### XY labels

```python
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("y=x^2");
plt.xlabel("X point")
plt.ylabel("Y point")
```

### Bold line

```python
plt.plot(x, y, "-o", lw=3)
plt.grid(True)
plt.title("y=x^2");
plt.xlabel("X point")
plt.ylabel("Y point")
```

### X log scale

```python
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("y=x^2");
plt.xlabel("X point")
plt.ylabel("Y point")
plt.xscale("log")
```

### Line colors

```python
for i in range(10):
    plt.plot(x[:5], y[:5]+i)
```

## imshow関連

### オリジナル画像

```python
img = Image.open("macaque.jpg")
img
```

### imshow

```python
img = Image.open("macaque.jpg")
img = np.asarray(img)
plt.imshow(img)
plt.title("macaque")
```

### imshow (plot overlay 1)

```python
img = Image.open("macaque.jpg")
img = np.asarray(img)
plt.imshow(img, extent=[0, 8, 0, 80], aspect="auto")
plt.plot(x, y, "-o")
plt.grid(True)
plt.title("macaque with y=x^2")
plt.gca().set(xlim=[-0.5, 9.5], ylim=[-5, 85]);
```

### imshow (plot overlay 2)

```python
img = Image.open("macaque.jpg")
img = np.asarray(img)
plt.imshow(img)
plt.plot(x * 60, y*4, "-or", lw=2)
plt.grid(True)
plt.title("macaque with y=x^2")
```
